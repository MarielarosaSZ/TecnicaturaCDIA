# üå°Ô∏è Mapas de calor con Seaborn
Seaborn ofrece la funci√≥n seaborn.heatmap a nivel de ejes para generar mapas de calor. Esta herramienta permite visualizar matrices de datos en dos dimensiones, representando los valores mediante una paleta de colores.

## üîß Uso b√°sico
En su forma m√°s simple, heatmap requiere como argumento un bloque de valores 2D. Si se utiliza un DataFrame, las etiquetas de filas y columnas se mostrar√°n autom√°ticamente en el gr√°fico.





```python
import seaborn as sns
import pandas as pd
import numpy as np

# Generaci√≥n de datos simulados
ventas = pd.DataFrame(
    np.random.randint(1, 100, 120).reshape(12, 10),
    columns = np.arange(2009, 2019),
    index = ["ene", "feb", "mar", "abr", "may", "jun", "jul", "ago", "sep", "oct", "nov", "dic"]
)

# Mapa de calor b√°sico
ax = sns.heatmap(ventas)
```

```python
# Personalizaci√≥n de colores : Puedes especificar una paleta de colores con el par√°metro cmap:
ax = sns.heatmap(ventas, cmap="PiYG")
```

```python
# Mostrar valores en las celdas: Para incluir los valores num√©ricos dentro de cada celda, usa el par√°metro annot:

ax = sns.heatmap(ventas, cmap="Blues", annot=True)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset (puede ser desde Drive o URL)
df = pd.read_csv('simpsons_episodes.csv')  # Asegurate de tener columnas: season, episode, imdb_rating
```

```python
df
```

```python
heatmap_data = df.pivot_table(values='imdb_rating', index='number_in_season', columns='season')
heatmap_data = heatmap_data.iloc[::-1]
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='RdYlGn', linewidths=0.5)
plt.title('Calificaciones IMDb por episodio - Los Simpsons')
plt.xlabel('Temporada')
plt.ylabel('Episodio')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Calificaciones IMDb por episodio - Los Simpsons')
plt.xlabel('Episodio')
plt.ylabel('Temporada')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Calificaciones IMDb por episodio - Los Simpsons')
plt.xlabel('Episodio')
plt.ylabel('Temporada')
plt.show()
```
